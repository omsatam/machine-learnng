# Linear Regression Diabetis dataset

**Importing required libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
diab=load_diabetes()
```

```python
diab.keys()
```

```python
diab2=pd.DataFrame(diab.data)
```

```python
diab2
```

```python
diab2.columns=diab.feature_names
```

```python
diab2
```

```python
diab2['target']=diab.target
```

```python
diab2
```

```python
diab.DESCR
```

```python
x=diab2[['age','sex','bmi','bp']]
```

```python
x
```

```python
y=diab2['target']
```

```python
y
```

```python
x_train,x_test,y_train,y_test=sklearn.model_selection.train_test_split(x,y,test_size=0.3,random_state=6)
```

```python
x_train.shape
```

```python
lm=LinearRegression()
y_train.shape
```

```python
lm.fit(x_train,y_train)
```

```python
y_pred=lm.predict(x_test)
```

```python
df=pd.DataFrame(y_pred,y_test)
```

```python
df
```

```python
plt.scatter(y_test,y_pred)
plt.plot(y_pred,lm.predict(x_test))
plt.xlabel='actual price'
plt.ylabel='predicted price'
plt.title('diabetis data')
plt.plot()
```

```python
mse=mean_squared_error(y_test,y_pred)
```
