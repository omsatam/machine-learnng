# Area plot

```python
days=[1,2,3,4,5]
sleep=[7,8,6,11,7]
eat=[2,3,4,3,2]
work=[7,8,7,2,2]
play=[8,5,7,8,13]
plt.plot([],[],color='m',label='sleep',linewidth=5)
plt.plot([],[],color='c',label='eat',linewidth=5)
plt.plot([],[],color='r',label='work',linewidth=5)
plt.plot([],[],color='k',label='play',linewidth=5)
plt.stackplot(days,sleep,eat,work,play,colors=['m','c','r','k'])
plt.xlabel('X')
plt.ylabel('Y')
plt.title('stack plot')
plt.legend()
plt.show()
```

# pie diagram

```python
slices=[7,2,5,14]
activities=['sleeping','eating','working','playing']
colors=['c','m','r','b']
plt.pie(slices,labels=activities,colors=colors,startangle=90,shadow=True,explode=(0,0.2,0,0),autopct='%1.1f%%')
plt.title('pie plot')
plt.show()
```

# Linear regression algorithm

**Importing required libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import sklearn
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
boston = load_boston()
boston.keys()
```

```python
bos=pd.DataFrame(boston.data)
bos
```

```python
bos.columns=boston.feature_names
bos
```

```python
bos['PRICE']=boston.target
```

```python
bos
```

**Selecting  dependent and independent variables**

```python
Y=bos['PRICE']
X=bos.drop('PRICE',axis=1)
```

```python
X
```

**Splitting data into training and testing with inbuilt train_test_split function**

```python
X_train,X_test,Y_train,Y_test=sklearn.model_selection.train_test_split(X,Y,test_size=0.3,random_state=6)
```

```python
X_train.shape
X_test.shape
```

```python
lm=LinearRegression()
lm.fit(X_train,Y_train)
```

```python
Y_test_pred = lm.predict(X_test)
```

```python
df=pd.DataFrame(Y_test_pred,Y_test)
df
```

```python
mse=mean_squared_error(Y_test,Y_test_pred)
mse
```

**plotting linear regression line and plot**

```python
Y_train_pred=lm.predict(X_train)
plt.scatter(Y_train,Y_train_pred,c='blue',marker='s',label='Training Data')
plt.scatter(Y_test,Y_test_pred,c='orange',marker='o',label='Testing Data')
plt.plot(Y_test_pred,lm.predict(X_test),label='Regression line')
plt.legend()
plt.xlabel=('True price')
plt.ylabel=('Predicted price')
plt.title('Linear Regression')
plt.plot()
```

**accuracy of the algorithm**

```python
lm.score(X_test,Y_test)
```
